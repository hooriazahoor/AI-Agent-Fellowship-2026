Interestingly, both prompts arrived at the same final number (23) here — but Output A gave no visible reasoning, which meant there was no way to check whether it actually calculated correctly or got lucky/pattern-matched a plausible-looking number.

Output B shows its work, which makes it possible to verify each step independently. In more complex multi-step problems (tested separately with a 3-step probability problem), the non-CoT prompt gave a wrong final answer, while the CoT version correctly caught its own arithmetic at each stage.

Chain-of-Thought is most valuable not just for getting a better answer, but for making the answer auditable — critical for anything where correctness matters (like scientific or medical use-cases).

# Improvements

For tasks where the reasoning itself doesn't need to be shown to the end user, a two-step "hidden CoT" approach works well: ask the model to reason step-by-step first, then in a follow-up instruction ask it to give only the final answer — combining accuracy with a clean user-facing response.

Explicitly asking the model to "double-check your arithmetic before giving the final answer" further reduced small calculation slips in testing.

# Observations

The zero-shot prompt strictly followed the three allowed categories (Positive/Negative/Neutral), landing on Neutral — a reasonable but slightly imprecise label, since the review clearly expresses both a positive point (fast delivery) and a negative one (damaged packaging).

The few-shot prompt, despite showing three clean single-sentiment examples, inferred a more nuanced label ("Mixed") because it picked up on the pattern of the task (sentiment classification of mixed opinions) rather than strictly the three literal labels shown.

This highlights an important lesson: few-shot examples teach the model the shape of the task, but the model may still generalize beyond the literal label set shown — which can be useful or risky depending on the use case.

# Improvements

To keep outputs strictly within the three intended categories, the prompt should explicitly state: "Only use one of these three labels: Positive, Negative, or Neutral — do not create new categories."
    
Adding one more few-shot example that specifically shows a mixed-opinion review being forced into one of the three categories would anchor the model's behavior more reliably.

Adding a fallback in code — stripping possible ``` fences and using try/except around json.loads() — makes the pipeline robust even on the rare occasions the model doesn't follow instructions exactly.

# Observations

Each added constraint (purpose → audience/length → tone/structure) produced a measurably more usable output, not just a stylistically different one.

The single biggest jump in quality came from Version 1 → 2, simply by adding a purpose and audience — vague prompts consistently produce vague, generic-sounding output because the model has nothing specific to anchor to.

Version 2 → 3 was a smaller but still meaningful jump — tone and structural instructions shifted how the content was delivered (relatable scene-setting vs. straightforward description) without changing what it was about.

# Improvements

A reusable "prompt checklist" emerged from this experiment — a good prompt should specify: (1) task, (2) audience, (3) format/length, (4) tone, (5) structure. Missing any of these leaves room for the model to guess, which is where generic output comes from.
    
For repeated use, this checklist could be turned into a template with placeholders — exactly the kind of prompt template built into the AI Workspace app in Assignment 3.
